```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image using OpenCV
image = cv2.imread('C:/Users/panca/OneDrive/Documents/praktikum_pcd/gambar1.png', cv2.IMREAD_GRAYSCALE)

# Ensure the image has been loaded
if image is None:
    print('Error: Could not open the image.')
    exit()

# Compute the histogram with 256 bins (for 8-bit images)
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

# Get the width and height of the image for normalization
height, width = image.shape

# Normalize the histogram by the image area
hist_norm = hist.ravel() / (width * height)

# Create a bar plot using Matplotlib
plt.figure()
plt.bar(range(256), hist_norm, align='center')
plt.xlim([0, 256])
plt.xlabel('Brightness Value')
plt.ylabel('Frequency')
plt.title('Brightness Histogram of the Image')

# Display the histogram
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# read the input image
image = cv2.imread('C:/Users/panca/OneDrive/Documents/praktikum_pcd/gambar1.png')

# define the alpha and beta values for contrast and brightness adjustment
alpha = 2.0
beta = 100

# adjust the contrast and brightness of the image
adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

# convert the adjusted image to grayscale
gray_image = cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2GRAY)

# apply histogram equalization to enhance the contrast
enhanced_grayscale = cv2.equalizeHist(gray_image)

# threshold the enhanced grayscale image to get a binary image
_, threshold_image = cv2.threshold(enhanced_grayscale, 150, 255, cv2.THRESH_BINARY)

# create a figure with 3 subplots
fig, axes = plt.subplots(1, 2, figsize=(8, 5))

# display the original, adjusted, and thresholded images in the subplots
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')

axes[1].imshow(threshold_image, cmap='gray')
axes[1].set_title('Conlvolved Image')

# display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

# Replace the file path with the path of your image file
file_path = 'C:/Users/panca/OneDrive/Documents/praktikum_pcd/gambar1.png'

# Load the image from the file path
image = mpimg.imread(file_path)

# Convert the image to grayscale
gray_image = image.mean(axis=2)

# Calculate the new dimensions of the image
new_height = 2 * image.shape[0]
new_width = 2 * image.shape[1]

# Create a new black image with the new dimensions
black_image = np.zeros((new_height, new_width))

# Copy the gray image into the bottom left corner of the new black image
start_row = image.shape[0]
start_col = 0
black_image[start_row:, start_col:start_col + image.shape[1]] = gray_image

# Display the new image using Matplotlib
plt.imshow(black_image, cmap='gray', extent=[0, new_width, 0, new_height])
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

# Replace the file path with the path of your image file
file_path = 'C:/Users/panca/OneDrive/Documents/praktikum_pcd/gambar1.png'

# Load the image from the file path
image = mpimg.imread(file_path)

# Convert the image to grayscale
gray_image = image.mean(axis=2)

# Mirror the gray image along the horizontal axis
mirrored_image = np.fliplr(gray_image)

# Calculate the new dimensions of the image
new_height = 2 * image.shape[0]
new_width = 2 * image.shape[1]

# Create a new black image with the new dimensions
black_image = np.zeros((new_height, new_width))

# Copy the mirrored image into the bottom right corner of the new black image
start_row = image.shape[0]
start_col = new_width - image.shape[1]
black_image[start_row:, start_col:] = mirrored_image

# Display the new image using Matplotlib
plt.imshow(black_image, cmap='gray', extent=[0, new_width, 0, new_height])
plt.show()
```
